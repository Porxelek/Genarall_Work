```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import requests
from tqdm import tqdm
import time
```

```python
def reader(file_path, sheet):
    try:
        df = pd.read_excel(file_path, sheet_name=sheet)
        df = df.dropna(thresh=9, axis=0)

    except Exception:
        df = pd.read_csv(file_path)
        df = df.dropna(thresh=9, axis=0)
    return df
```

```python
file_path = "C:\\Users\\madcat\\Desktop\\Scraping\\emine_hoca\\INVENTORY-HG (1).xlsx"
TOP50 = "TOP50"
SEC50 = "SEC50"
```

```python
top50 = reader(file_path, TOP50)
sec50 = reader(file_path, SEC50)
display(top50.head())
display(sec50.head())
```

```python
def link_controler(df, col):
    down = []
    up = []
    links = df[col]
    for index, link in enumerate(tqdm(links)):

        try:
            response = requests.get(link)
        except requests.exceptions.ConnectionError:
            down.append((index, link, time.ctime(time.time())))
            # print(link,": URL is down!")
        else:
            up.append((index, link, time.ctime(time.time())))
            # print(link,": URL is up!")

    return down, up
```

```python
def reporter(df, down, up):
    print(f"{len(down)} number of link(s) down")
    display(pd.DataFrame(down))
    print(f"{len(up)} number of link(s) up")
    display(pd.DataFrame(up, columns=["index", "Product Link", "Time"]))
    print(f"{len(down) / len(df) * 100}% of top 50 down")
    print(f"{len(up) / len(df) * 100}% of top 50 up")
```

```python
top50_down, top50_up = link_controler(top50, "Product Link")
sec50_down, sec50_up = link_controler(sec50, "Product Link")
```

```python
reporter(top50, top50_down, top50_up)
```

```python
reporter(sec50, sec50_down, sec50_up)
```
